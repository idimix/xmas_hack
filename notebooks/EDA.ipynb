```python
df = pd.read_pickle(f'{PATH_PROCESSED}/df.pickle')
mask_train = df['split'] == 'train'
mask_val = df['split'] == 'val'
mask_test = df['split'] == 'test'
```

```python
ax = df.groupby('month')['target'].mean().plot(style=':.')
ax.set_ylabel('target');
```

```python
ax = df.groupby('day_of_week')['target'].mean().plot(style=':.')
ax.set_ylabel('target');
```

```python
ax = df.query('is_working_day').groupby('day')['target'].mean().plot(style=':.', label='workday')
ax.set_ylabel('target')
ax.legend()

ax = df.query('~is_working_day').groupby('day')['target'].mean().plot(style=':.', label='weekend')
ax.set_ylabel('target')
ax.legend()
```

```python
ax = df.query('is_working_day').groupby('break_flight_start_hour')['target'].mean().plot(style=':.', label='workday')
ax.set_ylabel('target')
ax.legend()

ax = df.query('~is_working_day').groupby('break_flight_start_hour')['target'].mean().plot(style=':.', label='weekend')
ax.set_ylabel('target')
ax.legend()
```
