```python
df = pd.read_pickle(f'{PATH_PROCESSED}/df.pickle')
mask_train = df['split'] == 'train'
mask_val = df['split'] == 'val'
mask_test = df['split'] == 'test'
```

```python
def mape(y_train, predict):
    enumerator = np.abs(y_train - predict)
    denomirator = np.abs(y_train)
    mask = y_train != 0
    score = enumerator / denomirator
    score = score[mask]
    return np.mean(score)
```

```python
class MyCatBoostRegressorCV(BaseEstimator, RegressorMixin):
            
    def __init__(self, **params):
        self._init_params = params
        self.params = copy.deepcopy(params)
        self.cv = None
        self.models = []
        self.cat_features = []
        self.score = None
        self.scores = None
        self.predicts = None
        self.feature_names = None
        self.shap_values = []
        
    def get_params(self, deep=True):
        return self._init_params
    
    def set_params(self, **params):
        for param, value in params.items():
            self._init_params[param] = value
        self.__init__(**self._init_params)
        return self

    def fit(self, X, y):
        X = X.copy().reset_index(drop=True)
        y = y.copy().reset_index(drop=True)
        self.cat_features = X.columns[X.dtypes == 'O'].tolist()
    
        scores = []
        predict = []
        ids = []
        if self.cv is None:
            cv = KFold(n_splits=5, random_state=RANDOM_STATE, shuffle=True).split(X)
        else:
            cv = self.cv
        for ids_train, ids_test in cv:
            
            train = Pool(X.iloc[ids_train], y.iloc[ids_train], cat_features=self.cat_features)
            test = Pool(X.iloc[ids_test], y.iloc[ids_test], cat_features=self.cat_features)
            
            model = CatBoostRegressor(**self.params)
            model.fit(train, eval_set=test, verbose=False)

            predict_train = model.predict(train)
            predict_test = model.predict(test)
            score_train = mape(train.get_label(), predict_train)
            score_test = mape(test.get_label(), predict_test)
            scores.append(score_test)
            predict.append(predict_test)
            ids.append(ids_test)
            print('score_train:', score_train, 'score_test: ', score_test)

            self.models.append(copy.deepcopy(model))
            self.feature_names = train.get_feature_names()
            self.shap_values.append(model.get_feature_importance(test, type='ShapValues'))
            
        self.scores = scores
        self.score = np.mean(scores)
        ids = np.argsort(np.concatenate(ids))
        self.predicts = np.concatenate(predict)[ids]
        self.shap_values = np.vstack(self.shap_values)[ids]
        
        return self
       
    def predict(self, X):
        X = Pool(X, cat_features=self.cat_features)
        predict = []
        for model in self.models:
            predict.append(model.predict(X))
        predict = np.mean(predict, axis=0)
        return predict
    
    def get_feature_importance(self):
        return np.vstack([
            model.get_feature_importance() 
            for model in self.models
        ]).mean(axis=0)
    
    def get_feature_names(self):
        return self.feature_names
```

```python
features = ['target_break_flight_start_hour',
 'target_programme',
 'target_programme_prev',
 'tv_viewing',
 'target_break_distribution',
 'break_distribution',
 'target_programme_next',
 'day_of_week',
 'programme_age',
 'year_half',
 'is_working_day',
 'break_duration',
 'break_interval',
 'programme_flight_end_hour',
 'programme_num',
 'target_day_of_week',
 'day',
 'quarter',
 'break_content',
 'seconds_from_last_break',
 'programme_duration',
 'target_programme_category',
 'break_start_programme',
 'programme_flight_start_hour',
 'programme_category',
 'break_cnt',
 'break_end_programme',
 'programme_category_podkastlab',
 'target_programme_category_next',
 'break_distribution_pole_chudes',
 'programme_duration_real',
 'programme_category_next',
 'break_flight_end_hour',
 'target_programme_category_prev',
 'break_duration_all',
 'break_content_pole_chudes',
 'programme_duration_real_rate',
 'programme_prev',
 'programme_genre_next',
]

params = {
    'iterations': 3000,
    'learning_rate': 0.01,
    'task_type': 'GPU',
    'max_depth': 5,
    'random_seed': RANDOM_STATE,
    'loss_function': 'MAPE',
    "l2_leaf_reg": 15,
}

cv = [
    [df.query('month < 6').index.tolist(), df.query('month == 6').index.tolist()],
    [df.query('month < 7').index.tolist(), df.query('month == 7').index.tolist()],
    [df.query('month < 8').index.tolist(), df.query('month == 8').index.tolist()],
]

X_train = df.loc[mask_train | mask_val, features].reset_index(drop=True)
y_train = df.loc[mask_train | mask_val, 'target'].reset_index(drop=True)

model = MyCatBoostRegressorCV(**params)
model.cv = cv
model.fit(X_train, y_train)

print('cv:', np.mean(model.scores) + np.std(model.scores))
print('scores:', model.scores)
```

```python
# категориальные фичи
def target_encoding(df, mask, group_clms):
    feature = 'target_'+'_'.join(group_clms)
    if feature in df.columns:
        df = df.drop(feature, axis=1)
    
    group = df[mask].groupby(group_clms)
    target_sum = group['target'].transform('sum') - df[mask]['target']
    cnt = group['target'].transform('count') - 1
    target = pd.concat([df[mask_train][group_clms], (target_sum / cnt).to_frame(feature)], axis=1)
    target = target.groupby(group_clms, as_index=False).mean()
    # if 'month' in group_clms:
    #     target_month_1 = target.query('month == 1').copy()
    #     target['month'] += 1
    #     target = pd.concat([target, target_month_1], sort=False, ignore_index=True)
        
    df = df.merge(target, on=group_clms, how='left')
    
    # # заполняем пропуски
    # group_clms = [col for col in group_clms if col != 'month']
    # mask = df[feature].isna()
    # df.loc[mask, feature] = df.groupby(group_clms)[feature].transform('mean')
    return df

list_group_clms = [
    ['programme'],
    ['break_content'],
    ['break_distribution'],
    ['programme_genre'],
    ['programme_category'],
    ['break_flight_start_hour'],
    ['day_of_week'],
    ['programme_prev'],
    ['programme_genre_prev'],
    ['programme_category_prev'],
    ['programme_next'],
    ['programme_genre_next'],
    ['programme_category_next'],
]

for group_clms in list_group_clms:
    df = target_encoding(df, mask_train, group_clms)
```
